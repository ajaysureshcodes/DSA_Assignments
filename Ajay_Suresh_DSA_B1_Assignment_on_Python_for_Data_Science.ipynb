```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Renaming the data for better understanding based in the Iris flowers
df = df.rename(columns = {'SL':'Sepal Length', 'SW':'Sepal Width','PL':'Petal Length','PW':'Petal Width'})
```

```python
df.groupby('Classification')['Sepal Length'].sum().plot(kind = 'line', figsize= (10,6), marker = 'o', color = 'purple')
plt.xlabel('Classification')
plt.xticks(rotation = 0)
plt.ylabel('Average Sepal Length')
plt.title('Iris versions vs Avg Sepal Length')
plt.show()
```

```python
df.groupby('Classification')['Sepal Length'].plot.hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('Classification')['Petal Length'].mean().plot(kind='bar', color=['green', 'blue', 'coral'], edgecolor = 'black')
plt.title('Average Petal Length by Different Species in Iris Dataset')
plt.xlabel('Classification')
plt.ylabel('Average Petal Length')
plt.show()
```

```python
# Pair plot for feature comparison based on Classification
sns.pairplot(df, hue='Classification', markers=['o', 's', 'D'])
plt.title('Pair Plot for Iris')
plt.show()
```

```python
# Box plots for each Classification of Iris species
plt.figure(figsize=(12, 8))
sns.boxplot(x='Classification', y='Petal Width', data=df)
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

```python
# Create swarm plots for each species for Sepal Width
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Classification', y='Sepal Width', data=df, color = 'orange')
plt.title('Swarm Plot of Sepal Width by Species')
plt.show()
```

```python
# Correlation heatmap for all features
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features in Iris data')
plt.show()
```
