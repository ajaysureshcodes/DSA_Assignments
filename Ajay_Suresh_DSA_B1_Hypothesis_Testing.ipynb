1. **Clarifying if there is any increase in sales after
stepping into digital marketing.**

```python
#Checking the impact of digital marketing on sales using paired T-Test

t_stat, p_value = ttest_ind(df['Sales_before_digital_add(in $)'], df['Sales_After_digital_add(in $)'])

print("P Value:", p_value,"\n")

#Comparison of p-value with the significance level

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant increase in sales after implementing digital marketing.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in sales after implementing digital marketing.")
```

2. **Dependency between
the features “Region” and “Manager”.**

```python
##Checking dependency between Region and Manager using Chi-Square Test for Independence.

reg_man_table = pd.crosstab(df['Region'], df['Manager'])

chi2_stat, p_value, dof, exp = chi2_contingency(reg_man_table)

print("P Value:", p_value,"\n")

#Comparison of p-value with the significance level

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant dependency between Region and Manager.")
else:
    print("Fail to reject the null hypothesis. There is no significant dependency between Region and Manager.")
```
